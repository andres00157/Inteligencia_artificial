```python
import numpy as np
from sklearn import datasets
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.naive_bayes import BernoulliNB,GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import matthews_corrcoef,accuracy_score,f1_score
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

## Lectura datos
x,y = datasets.load_iris(return_X_y= True)
print(x.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)

F1 = []
mc = []
acc = []

for i in range(1,30):

  sc= StandardScaler()
  sc.fit(X_test)
  X_test=sc.transform(X_test)
  X_train=sc.transform(X_train)


  clf = BernoulliNB(alpha= i)
  clf.fit(X_train, y_train)

  y_predict= clf.predict(X_test)
  F1.append(f1_score(y_test, y_predict, average='micro'))
  mc.append(matthews_corrcoef(y_test, y_predict))
  acc.append(accuracy_score(y_test,y_predict))

plt.plot(F1)
plt.title("F1 score")
plt.show()
plt.plot(mc)
plt.title("matthews_corrcoef score")
plt.show()
plt.plot(acc)
plt.title("accuracy score")
plt.show()


pos= np.argmax(np.array(mc))

print("Mejor valor de alpha:",np.arange(1,30)[pos],"con matthews_corrcoef:",mc[pos],"con F1:",F1[pos],"con accuracy:",acc[pos])

```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)

F1 = []
mc = []
acc = []

for i in range(20):

  sc= StandardScaler()
  sc.fit(X_test)
  X_test=sc.transform(X_test)
  X_train=sc.transform(X_train)

  clf = GaussianNB(var_smoothing = 10**(-i/2))
  clf.fit(X_train, y_train)

  y_predict= clf.predict(X_test)
  F1.append(f1_score(y_test, y_predict, average='micro'))
  mc.append(matthews_corrcoef(y_test, y_predict))
  acc.append(accuracy_score(y_test,y_predict))

plt.plot(F1)
plt.title("F1 score")
plt.show()
plt.plot(mc)
plt.title("matthews_corrcoef score")
plt.show()
plt.plot(acc)
plt.title("accuracy score")
plt.show()

pos= np.argmax(np.array(mc))

print("Mejor valor de alpha:",10**(-np.arange(1,30)[pos]/2),"con matthews_corrcoef:",mc[pos],"con F1:",F1[pos],"con accuracy:",acc[pos])
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)

F1 = []
mc = []
acc = []

for i in range(10):

  sc= StandardScaler()
  sc.fit(X_test)
  X_test=sc.transform(X_test)
  X_train=sc.transform(X_train)

  clf = DecisionTreeClassifier(min_weight_fraction_leaf= i/20)
  clf.fit(X_train, y_train)

  y_predict= clf.predict(X_test)
  F1.append(f1_score(y_test, y_predict, average='micro'))
  mc.append(matthews_corrcoef(y_test, y_predict))
  acc.append(accuracy_score(y_test,y_predict))

plt.plot(F1)
plt.title("F1 score")
plt.show()
plt.plot(mc)
plt.title("matthews_corrcoef score")
plt.show()
plt.plot(acc)
plt.title("accuracy score")
plt.show()

pos= np.argmax(np.array(mc))

print("Mejor valor de alpha:",np.arange(1,30)[pos]/20,"con matthews_corrcoef:",mc[pos],"con F1:",F1[pos],"con accuracy:",acc[pos])
```
